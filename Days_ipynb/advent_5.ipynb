```python
import re

class Node:
    def __init__(self,prev=None,next=None,value=None):
        self.prev = prev
        self.next = next
        self.value = value
    
    def print_list(self):
        cur = self
        while cur:
            print(cur.value)
            cur = cur.next
    
    def find_end(self):
        cur = self
        while cur.next:
            cur = cur.next
        return cur
head = Node()
cur = head
seen = set()
for order in orders:
    first, second = re.findall("\d+|\d+",order)
    if first not in seen and second not in seen:
        seen.add(first)
        seen.add(second)
        cur.next = Node(prev=cur, value=first, next=None)
        cur.next.next = Node(prev=cur.next, value=second, next=None)
        cur = cur.next.next
        print(f"Order {first} and {second} added as new nodes.")
        head.print_list()
    
    elif second not in seen:
        seen.add(second)
        pointer = head
        while pointer.value != first:
            pointer = pointer.next
        
        temp = pointer.next
        insert_node = Node(prev=pointer, value=second, next=temp)
        pointer.next = insert_node
        
        cur = head.find_end()
        print(f"Order {first} and {second} added. Only second node was new.")
        head.print_list()
    
    elif  first not in seen:
        seen.add(first)
        pointer = head
        while pointer.value != second:
            pointer = pointer.next
        
        temp = pointer.next
        insert_node = Node(prev=pointer, value=first, next=temp)
        pointer.next = insert_node
        
        cur = head.find_end()
        print(f"Order {first} and {second} added. Only first node was new.")
        head.print_list()
    elif first in seen and second in seen:
        node_1 = head
        counter_1 = 0
        node_2 = head
        counter_2 = 0
        while node_1.value != first:
            node_1 = node_1.next
            counter_1 += 1
        while node_2.value != second:
            node_2 = node_2.next
            counter_2 += 1
        
        if counter_1 > counter_2:
            node_1.prev, node_1.next, node_2.prev, node_2.next = node_2.prev, node_2.next, node_1.prev, node_1.next
            if node_2.prev:
                node_2.prev.next = node_2
            if node_1.next:
                node_1.next.prev = node_1
                
        print(f"Order {first} and {second} swapped places.")
        head.print_list()
    else:
        print(f"Order {first} and {second} already in list.")
        head.print_list()

head.print_list()
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def find_node(self, value):
        current = self.head
        while current:
            if current.value == value:
                return current
            current = current.next
        return None

    def append(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    def add_values(self, left, right):
        left_node = self.find_node(left)
        right_node = self.find_node(right)

        if not left_node and not right_node:
            self.append(left)
            self.append(right)
        elif left_node and not right_node:
            new_node = Node(right)
            temp = left_node.next
            left_node.next = new_node
            new_node.prev = left_node
            new_node.next = temp
            if temp:
                temp.prev = new_node
        elif not left_node and right_node:
            new_node = Node(left)
            temp = right_node.prev
            right_node.prev = new_node
            new_node.next = right_node
            new_node.prev = temp
            if temp:
                temp.next = new_node
            else:
                self.head = new_node
        elif left_node and right_node:
            left_index = self.get_index(left_node)
            right_index = self.get_index(right_node)
            if right_index < left_index:
                self.swap_nodes(left_node, right_node)

    def get_index(self, node):
        current = self.head
        index = 0
        while current:
            if current == node:
                return index
            current = current.next
            index += 1
        return -1

    def swap_nodes(self, node1, node2):
        if node1 == node2:
            return

        # Swap the prev pointers
        if node1.prev:
            node1.prev.next = node2
        else:
            self.head = node2

        if node2.prev:
            node2.prev.next = node1
        else:
            self.head = node1

        node1.prev, node2.prev = node2.prev, node1.prev

        # Swap the next pointers
        if node1.next:
            node1.next.prev = node2

        if node2.next:
            node2.next.prev = node1

        node1.next, node2.next = node2.next, node1.next

    def print_list(self):
        current = self.head
        while current:
            print(current.value, end=" ")
            current = current.next
        print()
    
    def is_in_order(self, values):
        current = self.head
        for value in values:
            while current and current.value != value:
                current = current.next
            if not current:
                return False
            current = current.next
        return True

s = []

with open('advent_5_small_input.txt') as f:
    for line in f:
        s.append(line.strip())
sepearator = s.index("")

orders = s[:sepearator]
updates = s[sepearator+1:]
for i in range(len(updates)):
    updates[i] = updates[i].split(",")

dll = DoublyLinkedList()
for order in orders:
    left, right = re.findall("\d+|\d+", order)
    dll.add_values(left, right)

res = []
for update in updates:
    if dll.is_in_order(update):
        res.append(update)

sum = 0
for revision in res:
    sum += int(revision[len(revision)//2])


print(sum)
```
